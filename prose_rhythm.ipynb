```python
def info_corpus(corpusd):
    import pandas as pd
    keys=list(corpusd.keys())
    data=[corpusd[k].infod for k in keys]
    df=pd.DataFrame(data,index=keys)
    return df.sort_values('num_viols_per_10syll_avg')
```

```python
df=info_corpus(corpusd)
# df
```

```python
def plot_info_corpus(df,x='num_monosylls_per_10syll_avg',y='num_viols_per_10syll_avg'):
    import plotnine as p9
    fig=p9.ggplot(p9.aes(x=x,y=y,label='index'), data=df.reset_index())
    fig+=p9.geom_smooth(method='lm',alpha=.2,size=.5)
    
    
    fig+=p9.geom_text(size=8)
    fig+=p9.theme_classic()
    return fig



    
```

```python
plot_info_corpus(df)
```

```python
plot_info_corpus(df,y='num_parses_per_10syll_avg')
```
